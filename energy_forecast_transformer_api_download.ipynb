# Energy Consumption Forecast: Downloading Dataset via API

This notebook demonstrates how to download the Household Electric Power Consumption dataset using an API interface, process it, and prepare it for a machine learning project.

## 1. Using an API Interface to Download the Dataset

We check if the dataset can be downloaded via an API (e.g., Kaggle API, OpenML, UCI URL). The most common options are:
- **Kaggle API**: Allows downloading datasets after authentication.
- **OpenML API**: Offers many datasets, including energy consumption data.
- **Direct Download from UCI**: Via a public URL, usually without authentication.

The following describes the different options and prepares the download.

## 2. Authentication and API Access

### Kaggle API
To use the Kaggle API, you need an API key:
1. Create an account at [kaggle.com](https://www.kaggle.com/).
2. Go to "Account" and download your `kaggle.json` API key.
3. Save the file in the directory `~/.kaggle/` (Linux/Mac) or `C:\Users\<username>\.kaggle\` (Windows).

### OpenML API
For OpenML, no authentication is required for public datasets, but an API key can be helpful for larger requests.

### UCI Repository
The UCI Machine Learning Repository usually provides data without authentication via a direct URL.

```python
# 3. Data Retrieval and Storage
# We show three options: Kaggle API, OpenML API, and direct download from UCI

# --- Kaggle API (requires kaggle.json) ---
# !pip install kaggle
# import os
# os.environ['KAGGLE_CONFIG_DIR'] = os.path.expanduser('~/.kaggle')
# !kaggle datasets download -d uciml/electric-power-consumption-data-set -p ./data --unzip

# --- OpenML API ---
# !pip install openml
# import openml
# d = openml.datasets.get_dataset(151, download_all_files=True)  # 151 is the ID for the Household Power Consumption dataset
# df, *_ = d.get_data()
# df.to_csv('./data/household_power_consumption_openml.csv', index=False)

# --- UCI Repository (direct download) ---
import os
import requests
uci_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00235/household_power_consumption.txt'
os.makedirs('./data', exist_ok=True)
uci_path = './data/household_power_consumption.txt'
if not os.path.exists(uci_path):
    r = requests.get(uci_url)
    with open(uci_path, 'wb') as f:
        f.write(r.content)
print('Download complete!')
```
