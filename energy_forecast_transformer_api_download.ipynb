# Energieverbrauchsprognose: Datensatz per API herunterladen

Dieses Notebook zeigt, wie man den Household Electric Power Consumption Datensatz über eine API-Schnittstelle herunterlädt, verarbeitet und für ein Machine-Learning-Projekt vorbereitet.

## 1. API-Schnittstelle zum Herunterladen des Datensatzes verwenden

Wir prüfen, ob der Datensatz über eine API (z.B. Kaggle API, OpenML, UCI-URL) heruntergeladen werden kann. Die gängigsten Wege sind:
- **Kaggle API**: Ermöglicht das Herunterladen von Datensätzen nach Authentifizierung.
- **OpenML API**: Bietet viele Datensätze, darunter auch Energieverbrauchsdaten.
- **Direkter Download von UCI**: Über eine öffentliche URL, meist ohne Authentifizierung.

Im Folgenden werden die verschiedenen Möglichkeiten beschrieben und der Download vorbereitet.

## 2. Authentifizierung und API-Zugang

### Kaggle API
Um die Kaggle API zu nutzen, benötigst du einen API-Schlüssel:
1. Erstelle ein Konto auf [kaggle.com](https://www.kaggle.com/).
2. Gehe zu "Account" und lade deinen `kaggle.json` API-Schlüssel herunter.
3. Speichere die Datei im Verzeichnis `~/.kaggle/` (Linux/Mac) oder `C:\Users\<username>\.kaggle\` (Windows).

### OpenML API
Für OpenML ist keine Authentifizierung für öffentliche Datensätze nötig, aber ein API-Key kann für größere Anfragen hilfreich sein.

### UCI Repository
Das UCI Machine Learning Repository stellt die Daten meist ohne Authentifizierung über eine direkte URL bereit.

```python
# 3. Datenabruf und Speicherung
# Wir zeigen drei Varianten: Kaggle API, OpenML API und direkter Download von UCI

# --- Kaggle API (benötigt kaggle.json) ---
# !pip install kaggle
# import os
# os.environ['KAGGLE_CONFIG_DIR'] = os.path.expanduser('~/.kaggle')
# !kaggle datasets download -d uciml/electric-power-consumption-data-set -p ./data --unzip

# --- OpenML API ---
# !pip install openml
# import openml
# d = openml.datasets.get_dataset(151, download_all_files=True)  # 151 ist die ID für den Household Power Consumption Datensatz
# df, *_ = d.get_data()
# df.to_csv('./data/household_power_consumption_openml.csv', index=False)

# --- UCI Repository (direkter Download) ---
import os
import requests
uci_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00235/household_power_consumption.txt'
os.makedirs('./data', exist_ok=True)
uci_path = './data/household_power_consumption.txt'
if not os.path.exists(uci_path):
    r = requests.get(uci_url)
    with open(uci_path, 'wb') as f:
        f.write(r.content)
print('Download abgeschlossen!')
```
